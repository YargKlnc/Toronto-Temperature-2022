```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#setting x-axis (months)
x_axis_data = np.arange(1, 13, 1)
x_axis_data
```

```python
#average Toronto temperature, mean temperature data taken from climate.weather.gc.ca
yearly_temps_Celsius = [-8.3, -4.5, 1.1, 6.7, 15.8, 19.6, 22.7, 22.5, 17.8, 10.4, 5.4, -0.6]
```

```python
#plot the line
plt.plot(x_axis_data, yearly_temps_Celsius)
plt.show()
```

```python
# convert Celsius to Fahrenheit
yearly_temps_Fahrenheit = [((x * 9/5) + 32) for x in yearly_temps_Celsius]
yearly_temps_Fahrenheit
```

```python
# plot using Fahrenheit
plt.plot(x_axis_data, yearly_temps_Fahrenheit)
plt.show()
```

```python
# both C and F on same chart and giving axis labels
plt.plot(x_axis_data, yearly_temps_Celsius)
plt.plot(x_axis_data, yearly_temps_Fahrenheit)

plt.xlabel("Months of 2022")
plt.ylabel("Temperature (C in blue and F in orange)")

plt.show()
```

```python
plt.plot(x_axis_data, yearly_temps_Celsius)
plt.plot(x_axis_data, yearly_temps_Fahrenheit)

plt.xlabel("Months of 2022")
plt.ylabel("Temperature (C in green and F in orange)")

# Mark a blue circle to yearly_temps_Celsius
yearly_temps_Celsius_handle = plt.plot(x_axis_data, yearly_temps_Celsius, marker ='o', color='green', label="C")

# Mark a red triangle to yearly_temps_Fahrenheit
yearly_temps_Fahrenheit_handle = plt.plot(x_axis_data, yearly_temps_Fahrenheit, marker='^', color='orange', label="F")

# Adding a legend and setting on upper right and saving fig
plt.legend(loc="upper right")
plt.savefig("Toronto_Temps_2022.png")
plt.show()
```
